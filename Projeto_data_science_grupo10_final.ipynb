## **Descrição textual das variaveis de interesse principal.**

## Algumas informações desta seção podem se repetir mais a baixo onde escrevemos sobre as variaveis principais que usaremos na analise multivariada, vale ressaltar que após a analise univariada podem ser incluidas ideias que não estão presentes nesta seção.

A seguir, descrevemos as principais variáveis que julgamos que serão importantes:

**Idade (Age):**
Representa a idade do paciente mostrada em anos. A idade é um dos principais fatores de risco para diversas condições de saúde, incluindo doenças cardiovasculares, declínio cognitivo e síndromes presentes em pessoas idosas. E é frequentemente utilizada como variável de controle em estudos clínicos.

**Gênero (Gender):**
Esta variável indica o sexo biológico do individuo, categorizado como masculino ou feminino. O gênero é uma variável demográfica fundamental, pois diversas condições de saúde apresentam diferenças significativas entre homens e mulheres. Por exemplo, fatores hormonais, comportamentais e sociais podem influenciar o risco de doenças, desempenho cognitivo e até a resposta a tratamentos.

**Consumo de Álcool (AlcoholConsumption):**
Refere-se ao nível e frequência do consumo de bebidas alcoólicas. O consumo excessivo pode estar associado a efeitos adversos ligados a déficits cognitivos, enquanto níveis moderados podem ter efeitos controversos na saúde. Essa variável pode nos ajudar a entender se fatores comportamentais e de estilo de vida tem alguma influencia no diagnóstico.

**Tabagismo (Smoking):**
Indica se o indivíduo é fumante, ou não. O tabagismo é amplamente reconhecido como um fator de risco para uma série de doenças crônicas, incluindo câncer, doenças respiratórias, cardiovasculares e pode tambem acabar causando comprometimento cognitivo. A presença desse hábito pode potencializar riscos e afetar negativamente diversos indicadores clínicos e funcionais.

**MMSE (Mini-Exame do Estado Mental):**
O MMSE é um instrumento padronizado utilizado na avaliação do desempenho cognitivo global. O Score do MMSE auxilia na separação de déficits cognitivos leves a severos, sendo um dos principais indicadores utilizados em estudos sobre demência, Alzheimer e outras doenças neurodegenerativas.

**Avaliação Funcional (FunctionalAssessment):**
Esta variável mede a capacidade do indivíduo de realizar atividades cotidianas que requerem planejamento, coordenação e força física. Mostra o nível de independência e funcionalidade geral, sendo uma métrica importante para identificar estágios de fragilidade ou perda de autonomia.

**Queixas de Memória (MemoryComplaints):**
Representa relatos de problemas de memória, frequentemente utilizados como indicadores iniciais de comprometimento cognitivo leve. Essa variável pode ser considerada um preditor de risco para progressão a quadros mais graves de doenças cognitivas.

**Problemas Comportamentais (BehavioralProblems):**
Envolve alterações emocionais ou comportamentais, como agitação, agressividade, depressão ou apatia. Esses sintomas podem estar associados a condições neuropsiquiátricas ou neurológicas e impactam diretamente a qualidade de vida e o cuidado com o paciente. A detecção desses problemas pode auxiliar em intervenções mais eficazes.

**Atividades de Vida Diária (ADL – Activities of Daily Living):**
Refere-se à capacidade do indivíduo em realizar tarefas básicas de autocuidado, como alimentação, banho, vestimenta e locomoção. É um indicador essencial do grau de independência funcional e frequentemente utilizado para avaliar a progressão de doenças crônicas, síndromes geriátricas e comprometimento neurológico.

**Diagnóstico (Diagnosis):**
Representa a condição clínica do paciente em relação ao desenvolvimento da doença, podendo ser categorizada como positivo ou negativo. Esta é a variável target central na análise, pois define o grupo ao qual o paciente pertence em relação à presença do alzheimer.

```python
# calculo de valores de curtose, simetria e categorização de tipo de distribução, usamos o código para termos uma ideia mais precisa dos tipos.
# Selecionar as colunas numéricas, excluindo as colunas 'PatientID' e 'DoctorInCharge'
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_columns = [col for col in numeric_columns if col not in ['PatientID', 'DoctorInCharge']]

def distribution_type(skewness, kurt):
    if abs(kurt - 3) < 1:
        return "Distribuição Gaussiana (Normal)"
    elif kurt > 3:
        return "Distribuição Leptocúrtica (caudas pesadas)"
    elif kurt < 3:
        if skewness > 1:
            return "Distribuição Exponencial ou Log-normal (cauda à direita)"
        elif skewness < 1 and skewness > -1:
            return "Distribuição Uniforme (simétrica)"
        else:
            return "Distribuição com cauda à esquerda (negativa)"
    return "Distribuição Não Identificada"

for col in numeric_columns:
    print(f"\nVariável: {col}")

    skewness = skew(df_binario[col].dropna())
    kurt = kurtosis(df_binario[col].dropna())
    dist_type = distribution_type(skewness, kurt)

    print(f"Assimetria: {skewness:.2f}")
    print(f"Curtose: {kurt:.2f}")
    print(f"Tipo de distribuição: {dist_type}")
```

# Data Processing




**Essa seção esta organizada da seguinte maneira:**

O processamento de dados no geral, tanto padronização dos dados, separação em folds (feita nessa etapa para podermos demonstrar que o balanceamento foi feito apenas na divisão de treino) e balanceamento da variavél target, estão presentes nesta parte e abaixo de cada trecho do código que executa uma destas etapas existe uma célula de texto explicando e justificando o uso das técnicas.

```python
#normalização e encoding das variaveis categoricas

df['Diagnosis'] = LabelEncoder().fit_transform(df['Diagnosis'])  # 0 = negative, 1 = positive

X = df.drop(columns=['Diagnosis'])
y = df['Diagnosis']

# Definir colunas por tipo
numericas = [
    'Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality',
    'SleepQuality', 'SystolicBP', 'DiastolicBP', 'CholesterolTotal',
    'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides',
    'MMSE', 'FunctionalAssessment', 'ADL'
]

ordinais = ['EducationLevel']
ordem_educacional = [['None', 'High School', 'Bachelors', 'Higher']] #ordem

nominais = [
    'Gender', 'Ethnicity', 'Smoking', 'FamilyHistoryAlzheimers',
    'CardiovascularDisease', 'Diabetes', 'Depression', 'HeadInjury',
    'Hypertension', 'MemoryComplaints', 'BehavioralProblems', 'Confusion',
    'Disorientation', 'PersonalityChanges', 'DifficultyCompletingTasks',
    'Forgetfulness'
]

preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numericas), #normalização
    ('ord', OrdinalEncoder(categories=ordem_educacional), ordinais), #ordina encoding
    ('nom', OneHotEncoder(drop='first', sparse_output=False), nominais) # one hot encoding
])

X_processed = preprocessor.fit_transform(X)

nom_encoded = preprocessor.named_transformers_['nom'].get_feature_names_out(nominais)
column_names = numericas + ordinais + list(nom_encoded)

X_df = pd.DataFrame(X_processed, columns=column_names)
df_processed = pd.concat([X_df, pd.Series(y, name='Diagnosis')], axis=1)

df_processed.head()
```

Nesta parte foi realizado a normalização e codificação das variáveis.

As variáveis preditoras foram separadas por tipo: numéricas, ordinais e nominais. Para as variáveis numéricas, utilizou-se StandardScaler para padronizá-las e evitar que diferenças de escala interferissem na aprendizagem.

A variável ordinal EducationLevel foi codificada com OrdinalEncoder para preservar a ordem hierárquica entre os níveis educacionais.

As variáveis nominais foram transformadas com OneHotEncoder, convertendo categorias em multiplas colunas, nos permitindo representar dados categóricos sem induzir uma ordem artificial. Nesta etapa removemos a coluna da primeira categoria para evitar multicolinearidade, pois ela é assumida com base nas outras que estão presentes.

Essas transformações foram aplicadas com ColumnTransformer, que permite processar cada grupo de colunas com a técnica apropriada. No final os dados foram reunidos em um novo DataFrame apresentando um conjunto uniformemente numérico, codificado e pronto para a aplicação de modelos.

```python
#separação em k-fold (10)
# Separar X e y da base já normalizada e codificada
X = df_processed.drop(columns=['Diagnosis'])
y = df_processed['Diagnosis']

# Instanciar StratifiedKFold com 10 folds
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Criar DataFrame diretamente com list comprehension
shape_df = pd.DataFrame([
    {
        "Fold": f"Fold {fold}",
        "X_train_linhas": X.iloc[train_idx].shape[0],
        "X_train_colunas": X.iloc[train_idx].shape[1],
        "X_test_linhas": X.iloc[test_idx].shape[0],
        "X_test_colunas": X.iloc[test_idx].shape[1],
        "y_train_linhas": y.iloc[train_idx].shape[0],
        "y_test_linhas": y.iloc[test_idx].shape[0]
    }
    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), start=1)
])

# Exibir o DataFrame (opcional se estiver em notebook)
shape_df
```

**Divisão K-fold**

Nesta etapa, foi realizada a separação do conjunto de dados pré-processado em 10 subconjuntos de validação cruzada.

Fizemos esta etapa aqui para podermos mostrar que os balanceamentos que serão feitos mais pra frente foram feitos apenas na divisão de treino.

Primeiramente, a variável target Diagnosis foi separada das variáveis preditoras, mantendo a estrutura necessária para os testes.

Escolhemos o StratifiedKFold para garantir que cada fold mantenha a mesma proporção entre as classes da variável-alvo e mantenham os folds iguais para todos os modelos usando sempre o mesmo Random state (42).

Foi criado um DataFrame contendo as dimensões dos conjuntos de treino e teste de cada fold, permitindo demonstrar a consistência das divisões. Essa verificação mostra que todas as partições estejam completas e equilibradas.

```python
# chcando se as proporções dos diagnosticos se manteram
X = df_processed.drop(columns=['Diagnosis'])
y = df_processed['Diagnosis']

# Lista para armazenar proporções
proporcoes = []

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), start=1):
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    train_counts = y_train.value_counts(normalize=True).to_dict()
    test_counts = y_test.value_counts(normalize=True).to_dict()

    proporcoes.append({
        'Fold': f'Fold {fold}',
        'Train_Classe_0': train_counts.get(0, 0),
        'Train_Classe_1': train_counts.get(1, 0),
        'Test_Classe_0': test_counts.get(0, 0),
        'Test_Classe_1': test_counts.get(1, 0),
    })

df_proporcoes = pd.DataFrame(proporcoes)

df_proporcoes
```

**Verificação da distribuição:**

Nesta parte foi verificado se a proporção das classes da variável-alvo se manteve em cada divisão gerada pela validação cruzada estratificada.

Após aplicar o StratifiedKFold com 10 folds, foram calculadas, para cada partição, as proporções relativas das classes 0 (negativo) e 1 (positivo) tanto nos conjuntos de treino quanto de teste.

Essa verificação confirma a eficácia da estratificação, mostrando que a distribuição das classes se manteve próxima da original em todas as divisões.

O resultado foi organizado em um DataFrame, para facilitar a análise das proporções por fold.



```python
# Balanceamento da variavel target com 4 técnicas
X = df_processed.drop(columns=['Diagnosis'])
y = df_processed['Diagnosis']

# Armazenar resultados
contagens = []

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), start=1):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    # Aplicar balanceamentos
    X_smote, y_smote = SMOTE(random_state=42).fit_resample(X_train, y_train)
    X_ros, y_ros = RandomOverSampler(random_state=42).fit_resample(X_train, y_train)
    X_rus, y_rus = RandomUnderSampler(random_state=42).fit_resample(X_train, y_train)
    X_nm1, y_nm1 = NearMiss(version=1).fit_resample(X_train, y_train)

    # Tabela de contagens por classe
    df_counts = pd.concat([
        y_train.value_counts().rename("Treino").rename_axis("Classe"),
        y_test.value_counts().rename("Teste").rename_axis("Classe"),
        pd.Series(y_smote).value_counts().rename("SMOTE").rename_axis("Classe"),
        pd.Series(y_ros).value_counts().rename("RandomOver").rename_axis("Classe"),
        pd.Series(y_rus).value_counts().rename("RandomUnder").rename_axis("Classe"),
        pd.Series(y_nm1).value_counts().rename("NearMiss-1").rename_axis("Classe")
    ], axis=1).reset_index()

    df_counts["Fold"] = f"Fold {fold}"
    contagens.append(df_counts)

# Juntar todos os folds em um único DataFrame
df_resultado = pd.concat(contagens).set_index(["Fold", "Classe"])

display(df_resultado)
```

**Balanceamento:**

Nesta parte aplicamos o balanceamento da variável target "Diagnosis" nos dados de treino de cada fold da validação cruzada.

Após normalizar, separar os dados em 10 folds com StratifiedKFold e verificarmos as proporções, foram utilizadas quatro técnicas distintas para corrigir o desbalanceamento entre as classes qu tinham uma distribuição desbalanceada de 65% negativo e 35% positivo. Tomamos cuidado para sempre aplicar o balanceamento somente no conjunto de treino para não "trapacear" na avaliação do modelo.

**As técnicas aplicadas foram:**

SMOTE (Synthetic Minority Over-sampling Technique): Cria exemplos sintéticos da classe minoritária com base na interpolação de vizinhos próximos, aumentando sua representatividade sem replicar dados identicos.

RandomOverSampler: Duplica aleatoriamente amostras da minoria, sendo uma solução simples e eficaz, embora possa aumentar o risco de overfitting.

RandomUnderSampler: Reduz o número de exemplos da classe majoritária, promovendo equilíbrio com menor custo computacional, mas podendo perder informações relevantes.

NearMiss (versão 1): seleciona exemplos da classe majoritária que estão mais próximos da minoria, mantendo diversidade representativa.

Em cada um dos folds, foi registrada a contagem de exemplos por classe antes e depois de cada balanceamento.

Essas informações foram organizadas em um DataFrame, para comparar o efeito de cada técnica sobre a distribuição das classes.



# Machine Learning

**Essa seção está organizada da seguinte maneira:**

Os modelos de aprendizado de máquina utilizados no trabalho são apresentados abaixo, cada um dos 3 modelos foi treinado e testado em 10 Folds iguais e com os 4 tipos de balanceamento propostos acima no pré-processamento, depois de cada código de cada modelo existe uma breve descrição de como cada um foi construido.

```python
# lista dos balanceadores para facilitar no treino e teste com cada um deles
balanced_methods = {
    "RandomOver": RandomOverSampler(random_state=42),
    "SMOTE": SMOTE(random_state=42),
    "RandomUnder": RandomUnderSampler(random_state=42),
    "NearMiss-1": NearMiss(version=1)
}
```

```python
# knn usando os 4 tipos de balanceamento e o mesmo split com randomstate=42
results_knn = []

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]

    for bal_name, sampler in balanced_methods.items():
        X_bal, y_bal = sampler.fit_resample(X_train, y_train)

        model = KNeighborsClassifier(n_neighbors=5)
        model.fit(X_bal, y_bal)
        y_pred = model.predict(X_test)
        report = classification_report(y_test, y_pred, output_dict=True, zero_division=0)

        results_knn.append({
            "Fold": fold,
            "Balanceamento": bal_name,
            "Acurácia": round(accuracy_score(y_test, y_pred), 4),
            "Precision": round(report["1"]["precision"], 4),
            "Recall": round(report["1"]["recall"], 4),
            "F1-score": round(report["1"]["f1-score"], 4)
        })

df_knn = pd.DataFrame(results_knn)

df_knn
```

**KNN:**

Para cada uma das 10 dobras da validação cruzada estratificada foram aplicadas as diferentes técnicas de balanceamento aos dados de treino, ou seja cada dobra foi testada 4 vezes uma para cada tipo de balanceamento.

O modelo utilizado foi um classificador K-Nearest Neighbors com 5 vizinhos (k=5).

O desempenho foi avaliado nas métricas de acurácia, precisão, recall e F1-score com foco na classe positiva.


```python
# RF usando os 4 tipos de balanceamento e o mesmo split com randomstate=42
results_rf = []

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]

    for bal_name, sampler in balanced_methods.items():
        X_bal, y_bal = sampler.fit_resample(X_train, y_train)

# tunning do modelo
        model = RandomForestClassifier(
            n_estimators=200,
            max_depth=10,
            min_samples_split=5,
            min_samples_leaf=2,
            max_features='sqrt',
            class_weight='balanced',
            random_state=42
        )
        model.fit(X_bal, y_bal)
        y_pred = model.predict(X_test)
        report = classification_report(y_test, y_pred, output_dict=True, zero_division=0)

        results_rf.append({
            "Fold": fold,
            "Balanceamento": bal_name,
            "Acurácia": round(accuracy_score(y_test, y_pred), 4),
            "Precision": round(report["1"]["precision"], 4),
            "Recall": round(report["1"]["recall"], 4),
            "F1-score": round(report["1"]["f1-score"], 4)
        })

df_rf = pd.DataFrame(results_rf)
df_rf
```

# Evaluation metrics

**Explicação e justificativa das métricas:**

Nesta parte foram sintetizadas em um data frame as métricas de avaliação obtidas para cada modelo (KNN, Random Forest e SVM), considerando as diferentes técnicas de balanceamento aplicadas.

**Métricas utilizadas:**

A acurácia foi utilizada como métrica básica de avaliação por indicar a proporção total de acertos do modelo, considerando tanto as classificações corretas de diagnósticos positivos quanto de negativos.

A precision foi incluída para indicar quantos dos exemplos classificados como positivos estavam realmente corretos. Essa métrica é especialmente relevante quando se busca evitar falsos positivos, como no cenario de triagem médica.

O recall mede a capacidade do modelo de identificar corretamente todos os casos positivos reais. Em contextos clínicos, como no diagnóstico de Alzheimer, essa métrica é importante, pois prioriza a detecção de todos os casos verdadeiros da condição.

O F1-score foi utilizado por combinar precision e recall em uma única métrica, por meio da média harmônica. Ele é particularmente útil para ter um equilibrio entre evitar falsos positivos e falsos negativos.

O data frame final apresenta os dados do desempenho de cada modelo sob diferentes estratégias de balanceamento, servindo como base para análise comparativa.

```python
# Agrupar e calcular as médias por modelo e balanceamento
def agrupar_metricas(df, nome_modelo):
    metricas = ["Acurácia", "Precision", "Recall", "F1-score"]
    df_media = df.groupby("Balanceamento")[metricas].mean().round(4).reset_index()
    df_media["Modelo"] = nome_modelo
    return df_media

# Aplicar aos três modelos
df_knn_avg = agrupar_metricas(df_knn, "KNN")
df_rf_avg  = agrupar_metricas(df_rf, "RandomForest")
df_svm_avg = agrupar_metricas(df_svm, "SVM")

# Concatenar os três
df_metricas = pd.concat([df_knn_avg, df_rf_avg, df_svm_avg], ignore_index=True)

# Definir a ordem dos balanceamentos
ordem_balanceamentos = ["RandomOver", "SMOTE", "RandomUnder", "NearMiss-1"]
df_metricas["Balanceamento"] = pd.Categorical(df_metricas["Balanceamento"],
                                              categories=ordem_balanceamentos,
                                              ordered=True)

# Ordenar por modelo e balanceamento
df_metricas = df_metricas.sort_values(by=["Modelo", "Balanceamento"])

# Criar coluna combinada
df_metricas["Modelo_Balanceamento"] = df_metricas["Modelo"] + " - " + df_metricas["Balanceamento"].astype(str)

# Reorganizar colunas como no df_acuracia_final, mas com todas as métricas
df_final = df_metricas[["Modelo_Balanceamento", "Acurácia", "Precision", "Recall", "F1-score"]].reset_index(drop=True)

# Exibir
df_final
```

# Validation

**Essa seção está organizada da seguinte maneira:**

Primeiro apresentamos um texto explicando o protocolo de validação usado, depois os testes de hipótese relacionados aos modelos testados na seção anterior estão demonstrados aqui, abaixo do código que executa uma etapa do teste tem uma descrição demonstrando o que foi feito, e no final existe uma conclusão sobre as diferenças estatisticas dos modelos.

**Protocolo de validação**

Para avaliar o desempenho dos modelos de forma consistente, foi utilizado o protocolo de validação cruzada estratificada com 10 folds, implementado nas etapas anteriores. Essa abordagem garante que cada subdivisão dos dados preserve a proporção original das classes, reduzindo o viés na avaliação. Além disso, foi utilizado o mesmo valor de random_state=42 para assegurar que todos os modelos fossem testados com as mesmas partições, permitindo comparações justas e reproduzíveis entre eles.

Cada um dos 3 modelos foi treinado com os 4 tipos de balanceamento e testados/validados em conjutos de testes que não foram balanceados (seguindo a proporção padrão da variavél target presente no dataset original), depois disso foram obtidas as métricas citadas na célula acima.

Com o objetivo de validar estatisticamente as diferenças de desempenho entre os modelos e combinações de balanceamento, foi aplicado o teste de Friedman, que é adequado para a comparação de múltiplos modelos sobre amostras pareadas. O teste indicou que existem diferenças estatisticamente significativas, e, em seguida, foi realizado um teste post-hoc, utilizando a biblioteca autorank, que permitiu visualizar e quantificar as distâncias estatísticas entre os modelos.

Essa análise reforçou a confiabilidade dos resultados obtidos, garantindo que as conclusões não se baseassem apenas em métricas, mas também em evidências estatísticas robustas.

```python
# Abreviações de técnicas de balanceamento
abreviacoes_balanceamento = {
    "RandomOver": "ROS",
    "SMOTE": "SMT",
    "RandomUnder": "RUS",
    "NearMiss-1": "NM1"
}

# Abreviações de nomes de modelo
abreviacoes_modelo = {
    "KNN": "KNN",
    "RandomForest": "RF",
    "SVM": "SVM"
}

# Aplicar as abreviações
df_knn["Modelo"] = "KNN"
df_rf["Modelo"] = "RandomForest"
df_svm["Modelo"] = "SVM"

df_all = pd.concat([df_knn, df_rf, df_svm], ignore_index=True)
df_all["Modelo"] = df_all["Modelo"].map(abreviacoes_modelo)
df_all["Balanceamento"] = df_all["Balanceamento"].map(abreviacoes_balanceamento)

# Criar coluna combinada
df_all["Modelo_Balanceamento"] = df_all["Modelo"] + "-" + df_all["Balanceamento"]

# Pivotar
df_wide = df_all.pivot(index="Fold", columns="Modelo_Balanceamento", values="Acurácia").round(4)

styled_table = (
    df_wide.style
    .set_caption("Acurácias por Fold e por Modelo + Balanceamento")
    .set_table_styles(
        [{'selector': 'th', 'props': [('text-align', 'center')]},
         {'selector': 'td', 'props': [('text-align', 'center')]}]
    )
    .format("{:.4f}")
)
styled_table
```

**Tabela de acuracia dos modelos por fold:**

Nesta etapa, os resultados obtidos nos experimentos com validação cruzada feitos acima foram organizados para permitir uma análise estatística comparativa entre os modelos e para organiza-los para as próximas etapas.

Para isso, foram avaliadas 12 combinações entre três classificadores – KNN, Random Forest (RF) e SVM – e quatro técnicas de balanceamento de dados, SMOTE (SMT), Random UnderSampling (RUS), Random OverSampling (ROS) e NearMiss-1 (NM1).

Cada entrada do data frame representa a acurácia obtida por uma combinação específica em um dos folds da validação cruzada (com n=10 folds).

Foi criada uma coluna com a junção de modelo e balanceamento (Modelo-Balanceamento) e os dados foram pivotados, ficando com os folds como linhas e as 12 combinações como colunas.

```python
# Aplicar o teste para todas as 12 colunas de acurácia (uma por modelo + balanceamento)
stat, p = friedmanchisquare(
    df_wide["KNN-ROS"],
    df_wide["KNN-SMT"],
    df_wide["KNN-RUS"],
    df_wide["KNN-NM1"],
    df_wide["RF-ROS"],
    df_wide["RF-SMT"],
    df_wide["RF-RUS"],
    df_wide["RF-NM1"],
    df_wide["SVM-ROS"],
    df_wide["SVM-SMT"],
    df_wide["SVM-RUS"],
    df_wide["SVM-NM1"]
)

# Interpretar o resultado
if p < 0.05:
    print("Há diferença estatística entre os classificadores")
else:
    print("Não há diferença estatística entre os classificadores")
```

**Teste de Friedman:**

Com os dados organizados, foi aplicado o teste de Friedman, um teste estatístico não paramétrico e pareado indicado para comparar três ou mais métodos em testes com mesmos dados onde não há uma distribuição normal (que é exatamente o caso deste trabalho).

A ideia do teste é verificar se existe diferença estatisticamente significativa entre os métodos avaliados. A formulação das hipoteses do teste são:

H₀ (hipótese nula): Não há diferença entre os métodos (todos têm desempenho equivalente).

H₁ (hipótese alternativa): Ao menos um método difere significativamente dos demais.

O teste foi aplicado nas 12 colunas de acurácia da tabela. Se o p-valor resultante for menor que 0.05, rejeita-se H₀ e conclui-se que há diferença significativa entre os métodos.

Neste experimento, o p-valor obtido foi inferior ao limite, o que justifica a realização de testes post-hoc para identificar quais métodos diferem entre si.

```python
# Aplicar o autorank usando a tabela de acurácia por fold e modelo + balanceamento
ranked_results = autorank(
    df_wide,                      # DataFrame com folds como linhas e classificadores como colunas
    order='descending',          # Quanto maior a acurácia, melhor
    alpha=0.01,                  # Nível de significância (95%)
    force_mode='nonparametric', # Força o uso de testes não paramétricos
    verbose=False                # Suprimir saída intermediária
)

# Exibir gráfico com ranking dos modelos
plot_stats(ranked_results)
```

**Teste Post-hoc Nemenyi:**

Após a rejeição da hipótese nula, foi aplicado o módulo autorank, que realiza testes post-hoc baseados em rankings médios entre os métodos.

Este gráfico exibe visualmente a ordem dos métodos com base em seu desempenho médio. Linhas horizontais conectam os métodos que não apresentam diferença estatisticamente significativa entre si, conforme o teste post-hoc.
